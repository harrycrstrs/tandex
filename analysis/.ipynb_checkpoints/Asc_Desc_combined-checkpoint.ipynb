```python
import xarray as xr
import os 
import glob
from matplotlib import pyplot as plt
import numpy as np
import rioxarray as rx
import geopandas as gpd
from datetime import datetime as dt
import pandas as pd
from mpl_toolkits.axes_grid1 import AxesGrid
from mpl_toolkits.axes_grid1 import ImageGrid
from scipy import stats

# Get TDX derived DEMs ------------------------------------
DEM_dir = '/disk/scratch/local.4/harry/DEM/NC/merged/subset/'
files = glob.glob(DEM_dir+'*.nc')
f_per = [f for f in files if 'PER' in f]
f_gab = [f for f in files if 'GAB' in f]
f_gabA = [f for f in f_gab if '_A_' in f]
f_gabD = [f for f in f_gab if '_D_' in f]

# Get forest inventory data --------------------------------------------------------
INVENTORY_dir='/exports/csce/datastore/geos/groups/MitchardGroupData/FODEX/Forest_inventory/'
g_df = pd.read_csv(INVENTORY_dir+'GabonPlots.csv',index_col='Plot')
p_df = pd.read_csv(INVENTORY_dir+'PeruPlots.csv',index_col='Plot')

# Get plot shapefiles -------------------------------------------
SHP_dir='/exports/csce/datastore/geos/groups/MitchardGroupData/FODEX/core_plot_shapefiles/'
shps = gpd.read_file(SHP_dir+'core_plots_10mBuffer.shp')
#shps_nobuff = gpd.read_file(SHP_dir+'core_plots.shp')
shps_bg = gpd.read_file(SHP_dir+'background_plots.shp')
```

```python
# Convenience routines ----------

def open_ds(dataset,looks):
    # Open net_cdf file ready to be cropped using rio
    filename = [x for x in dataset if '_ML'+str(looks)+'_' in x][0]
    try:
        ds = xr.open_dataset(filename).drop(['metadata','crs']).rio.write_crs('EPSG:4326')
    except:
        ds = xr.open_dataset(filename).rio.write_crs('EPSG:4326')
    return ds

def crop(ds,plot,all_touched=False):
    # Crop an xarray object 'ds' to plot
    # plot is a string e.g. 'PC1' or 'GC4'
    # pass all_touched = True to include pixels that partially intersect the plot
    ds.rio.set_spatial_dims(x_dim='lon', y_dim='lat', inplace=True)
    if 'B' in plot:
        shape = shps_bg[shps_bg.layer==plot].geometry
    else:
        shape = shps[shps.layer==plot].geometry
    return ds.rio.clip(shape,all_touched=all_touched)
```

```python
# open ML3 datasets
ascending = open_ds(f_gabA,3)
descending = open_ds(f_gabD,3).interp(lat=ascending.lat,lon=ascending.lon,method='nearest')
```

```python
def changeMap(asc,des):
#coherent
    cohA = asc.coh.median(dim='t')
    cohD = des.coh.median(dim='t')
    whereA = cohA > cohD
    neither = (cohA < 0.5) & (cohD < 0.5)
    changeA = asc.height.isel(t=[2,3]).mean(dim='t') - asc.height.isel(t=[0,1]).mean(dim='t')
    changeD = des.height.isel(t=[2,3]).mean(dim='t') - des.height.isel(t=[0,1]).mean(dim='t')
    change = changeA.where(whereA,other=changeD)
    change = xr.Dataset(data_vars={'coherent':change - change.mean()})
#naive all 8
    ds = xr.concat([ascending,descending],dim='t').sortby('t')
    before = ds.height.isel(t=[0,1,2,3]).mean(dim='t')
    after = ds.height.isel(t=[4,5,6,7]).mean(dim='t')
    c = after - before
    change['naive8'] = c-c.mean()
#naive 4s
    for i in [0,1]:
        for j in [2,3]:
            for k in [4,6]:
                for l in [5,7]:
                    before = ds.height.isel(t=[i,j]).mean(dim='t')
                    after = ds.height.isel(t=[k,l]).mean(dim='t')
                    label = 'naive'+str(i)+str(j)+str(k)+str(l)
                    c = after - before
                    change[label]=c-c.mean()
#ascending
    before = ds.height.isel(t=[2,3]).mean(dim='t')
    after = ds.height.isel(t=[0,2]).mean(dim='t')
    c = after - before
    change['ascending'] = c-c.mean()
#descending
    before = ds.height.isel(t=[0,1]).mean(dim='t')
    after = ds.height.isel(t=[1,3]).mean(dim='t')
    c = after - before
    change['descending'] = c-c.mean()
    return change
```

```python
def changeMap(LOOKS,passdir='both',plotting=False):
    if passdir=='both':
        ascending = open_ds(f_gabA,LOOKS)
        descending = open_ds(f_gabD,LOOKS).interp(lat=ascending.lat,lon=ascending.lon,method='nearest')
        changeMap = get_coh_changeMap(ascending,descending)
    elif passdir=='A':
        ds = open_ds(f_gabA,LOOKS)
        changeMap = ds.height.isel(t=[2,3]).mean(dim='t') - ds.height.isel(t=[0,1]).mean(dim='t')
    elif passdir=='D':
        ds = open_ds(f_gabD,LOOKS)
        changeMap = ds.height.isel(t=[2,3]).mean(dim='t') - ds.height.isel(t=[0,1]).mean(dim='t')
    q = changeMap.mean()
    DH_list= []
    DAGB_list = []

    for plot in ['GC'+i for i in ['1','2','3','4']]:
        DH = crop(changeMap,plot,all_touched=True).mean() - q
        DAGB = -g_df['AGB_loss_pct'][plot]
        DAGB_err = g_df['dAGB_loss_pct'][plot]
        if plotting == True:
            plt.plot(DAGB,DH,'s',label=plot,color='black')
            plt.errorbar(DAGB,DH,xerr=DAGB_err,color='black')
            plt.annotate(plot,xy=(DAGB-0.5,DH+0.1))

        DH_list.append(DH)
        DAGB_list.append(DAGB)

    slope, intercept, r, p, se = stats.linregress(DAGB_list,DH_list)
    def best_fit(dagb):
        return intercept + slope*dagb
    if plotting == True:
        plt.plot(np.linspace(-25,5),best_fit(np.linspace(-25,5)),color='grey')
    DHmin = 999
    for plot in ['GB'+i for i in ['1','2','3','4','5','6']]:
        DH = crop(changeMap,plot,all_touched=True).mean() -q
        DAGB = 0
        if plotting == True:
            plt.plot(DAGB,DH,'o',label=plot)
            plt.annotate(plot,xy=(DAGB-2,DH))
        if DH < DHmin:
            DHmin = DH
    sensitivity = (DHmin - intercept) / slope
    
    if plotting ==True:
        plt.axvline(x=sensitivity,linestyle='--',color='grey')
        plt.axhline(y=0,color='grey')
        plt.xlabel('$\Delta$AGB (%)')
        plt.ylabel('$\Delta h_\phi$ (m)')
        plt.title('Combined Ascending and Descending Passes')
        #plt.savefig('/home/s1332488/chapter1/dH_vs_dAGB_'+str(LOOKS)+'Looks_AscDesCombined.png',dpi=300)
        plt.show()    
    return pd.Series({'looks':LOOKS,
                      'pass':passdir,
                      'slope':slope,
                      'r': r, 
                      'sensitivity':float(sensitivity),
                      'p':p, 
                      'se':se})
```

```python
df_stats = pd.DataFrame(columns=['looks','pass','slope','r','sensitivity','p','se'])
for L in [int(f.split('ML')[1].split('_')[0]) for f in f_gabD]:
    for P in ['A','D','both']:
        df_stats = df_stats.append(changeMap(L,passdir=P),ignore_index=True)
```
