```python
# -*- coding: utf-8 -*-
"""
Created on Thu May  6 16:37:41 2021

@author: s1326314
"""

import numpy as np
import rasterio as rs
import geopandas as gpd
from rasterio.features import shapes
from rasterio import features
import xarray as xr

# transform raster into grid of pixels
def geopandasFromRaster(infile,crs,n):
    """
    infile: raster file
    crs: coordinate system
    n: pixel area
    """
    mask = None
    with rs.Env():
        with rs.open(infile) as src:
            #get grid of pixels with their raster value
            image = src.read(4) #choose nir band
            results = (
            {'properties': {'raster_val': v}, 'geometry': s}
            for i, (s, v) 
            in enumerate(
                shapes(image, mask=mask, transform=src.transform)))
            
    geoms = list(results)
    df = gpd.GeoDataFrame.from_features(geoms,crs)
    #add area of each pixel
    df["area"] = df['geometry'].area
   
    df["weight"] = df["area"]/n
    # check for weights > 1
    exc = df.loc[(df.weight > 1)]
    # check if any data has weight > 1
    print('weights > 1: ' +str(len(exc)) +' for '+ infile)
    print(exc.weight)
    
    #convert to vector
    gdf = gpd.GeoDataFrame(df, geometry = df.geometry)
    #gdf.to_file("Analysis/Optical/Peru/materials/weights/weight_raster.shp", driver = 'ESRI Shapefile')
    return gdf

# convert grid of weighted pixel into raster with weight band
def geopandasToRaster(infile,out_fn,crs,n):
    """
    infile: raster file
    outfn: raster outfile
    crs: coordinate system
    n: pixel area
    """    
    rts = rs.open(infile)
    gdf = geopandasFromRaster(infile,crs,n)

    #get metadata from input file
    meta = rts.meta.copy()
    meta.update(compress='lzw')
    meta.update(dtype='float32')
    with rs.open(out_fn, 'w+', **meta) as out:
        out_arr = out.read(1)

        # this is where we create a generator of geom, value pairs to use in rasterizing
        shapes = ((geom,value) for geom, value in zip(gdf.geometry, gdf.weight))

        burned = features.rasterize(shapes=shapes, fill=0, out=out_arr, transform=out.transform)
        out.write_band(1, burned)

    return
```
