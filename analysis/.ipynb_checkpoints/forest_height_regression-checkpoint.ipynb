## Attempt at RF regression to get forest height from TDX variables
### 21st July 2021: had a go using EE to resample Canopy height from UAV data and TDX (3looks) phase height/ coherence metrics to 50/100 m. No sign of any great results

```python
from sklearn.ensemble import RandomForestRegressor as rfr
from sklearn.model_selection import train_test_split
```

```python
import rasterio as rs
import numpy as np
import os 
import glob
from matplotlib import pyplot as plt
```

```python
os.chdir('/exports/csce/datastore/geos/users/s1332488/chapter1_data/forest_height/')
glob.glob('*')
```

```python
def read_in(datasets,band_numbers):
    features = []
    for B in band_numbers:
        features.append(np.concatenate( [D.read(B).flatten() for D in datasets] , axis =0))
    return np.transpose(features)
```

```python
dataG = rs.open('50m_TDXfeatures_vs_forest_height_gabon.tif')
dataP = rs.open('50m_TDXfeatures_vs_forest_height_peru.tif')
X = read_in([dataG],[1,2,3,4,5])
Y = read_in([dataG],[6])

valued = ~np.isnan(Y).flatten()
Y = Y[valued,:].ravel()
X = X[valued,:]

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.25, random_state=21)

print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

```python
regr = rfr(n_estimators=25,max_depth=3,random_state=0)
regr.fit(X_train,y_train)
```

```python
print(regr.score(X_test,y_test))
prediction = regr.predict(X_test)
plt.plot(y_test,prediction,'.')
```

```python
plt.plot(X[:,1],X[:,2],'.')
```

```python
DD = [2,3,4,5,6,7,8,9,10]
for N in [2,5,10,25,50,100]:
    R = []
    for D in DD:
        regr = rfr(n_estimators=N,max_depth=D,random_state=0)
        regr.fit(X_train,y_train)
        R.append(regr.score(X_test,y_test))
    plt.plot(DD,R,label=str(N))
plt.legend() 
plt.show()
```
