```python
import xarray as xr
import os 
import glob
from matplotlib import pyplot as plt
import numpy as np
import rioxarray as rx
import geopandas as gpd
from datetime import datetime as dt
import pandas as pd
from mpl_toolkits.axes_grid1 import AxesGrid
from mpl_toolkits.axes_grid1 import ImageGrid
from scipy import stats

# Get TDX derived DEMs ------------------------------------
DEM_dir = '/disk/scratch/local.4/harry/DEM/NC/merged/subset/'
files = glob.glob(DEM_dir+'*.nc')
f_per = [f for f in files if 'PER' in f]
f_gab = [f for f in files if 'GAB' in f]
f_gabA = [f for f in f_gab if '_A_' in f]
f_gabD = [f for f in f_gab if '_D_' in f]

# Get forest inventory data --------------------------------------------------------
INVENTORY_dir='/exports/csce/datastore/geos/groups/MitchardGroupData/FODEX/Forest_inventory/'
g_df = pd.read_csv(INVENTORY_dir+'GabonPlots.csv',index_col='Plot')
p_df = pd.read_csv(INVENTORY_dir+'PeruPlots.csv',index_col='Plot')

# Get plot shapefiles with 10 metre buffer-------------------------------------------
SHP_dir='/exports/csce/datastore/geos/groups/MitchardGroupData/FODEX/core_plot_shapefiles/'
shps = gpd.read_file(SHP_dir+'core_plots_10mBuffer.shp')
shps_nobuff = gpd.read_file(SHP_dir+'core_plots.shp')
shps_bg = gpd.read_file(SHP_dir+'background_plots.shp')
```

```python
# Convenience routines ----------

def open_ds(filename):
    # Open net_cdf file ready to be cropped using rio
    try:
        ds = xr.open_dataset(filename).drop(['metadata','crs']).rio.write_crs('EPSG:4326')
    except:
        ds = xr.open_dataset(filename).rio.write_crs('EPSG:4326')
    return ds

def crop(ds,plot,all_touched=False):
    # Crop an xarray object 'ds' to plot
    # plot is a string e.g. 'PC1' or 'GC4'
    # pass all_touched = True to include pixels that partially intersect the plot
    ds.rio.set_spatial_dims(x_dim='lon', y_dim='lat', inplace=True)
    if 'B' in plot:
        shape = shps_bg[shps_bg.layer==plot].geometry
    else:
        shape = shps[shps.layer==plot].geometry
    return ds.rio.clip(shape,all_touched=all_touched)
```

```python
logging = {'GC1':['2020-01-24','2020-01-27'],
           'GC2':['2020-01-27'],
           'GC3':['2020-01-24','2020-02-04'],
           'GC4': ['2020-01-28'] ,
           'PC1': ['2019-07-29'],
           'PC2': ['2019-07-29'],
           'PC3': ['2019-07-29'],
           'PC4': ['2019-07-29'] 
          }
```

```python
def get_reference(f,country):
    if country == 'P':
        ref = [f for f in f_per if 'ML3_' in f][0]
    elif country == 'G':
        if '_A_' in f:
            ref = [f for f in f_gabA if 'ML3_' in f][0]
        elif '_D_' in f:
            ref = [f for f in f_gabD if 'ML3_' in f][0]
    return ref

def get_phase_height_change(f,offset=True,show=False,coh_threshold=0.45):
    ds = open_ds(f)
    looks = int( f.split('_ML')[1].split('_')[0] )
    print('Looks: '+str(looks))
    if 'GAB' in f:
        country = 'G'
        df = g_df
    else:
        country = 'P'
        df = p_df
    plotnames = [country+p for p in ['C1','C2','C3','C4']] + [country+'B'+n for n in ['1','2','3','4','5']]
    print(plotnames)
    if offset==True:
        offset = ds.height.mean(dim=['lat','lon'])
        print('Offset: ')
        print(offset.values)
        ds['height'] = (ds.height - offset)
    elif offset=='coherent':
        offset = ds.height.where( ds.coh > ds.coh.quantile(0.9,dim=['lat','lon']) ).median(dim=['lat','lon'])
        print('Coherent offset: ')
        print(offset.values)
        ds['height'] = (ds.height - offset).values
    with open_ds(get_reference(f,country)) as ref:
        ds = ds.interp_like(ref,method='nearest')
    results = {}
    for P in plotnames:
        # Extract one hectare plot
        timeseries = crop(ds,P)
        # Getting date when logging occurred 
        try:
            logdate = logging.get(P)[0]
        except:
            if country == 'P':
                logdate = '2019-07-29'
            else:
                logdate = '2020-01-24'
        # compare before and after logging       
        before = timeseries.sel(t=slice(None,logdate)).height
        after = timeseries.sel(t=slice(logdate,None)).height
        difference = after.mean(dim='t') - before.mean(dim='t')
        
        if show==True:
            ax = plt.axes()
            difference.plot(ax=ax,center=0)
            ax.set_title(P)
            if 'C' in P:
                shps_nobuff[shps_nobuff.layer==P].boundary.plot(ax=ax,color='red')
            else:
                shps_bg[shps_bg.layer==P].boundary.plot(ax=ax,color='red')
            plt.show()
            
        if coh_threshold != None:
            timeseries.coh.mean(dim='t').plot()
            plt.show()
            difference = difference.where(timeseries.coh.mean(dim='t') > coh_threshold)
            difference = np.nanmean( difference.values )
        else:
            difference = float(difference.mean().values)
        results[P] = difference
    return pd.Series(data=results)
```

```python
def plot_DH_vs_dAGB(tdx_data,plot_df,plot_shapes,
                    looks=['4','8','16','32'],offset=True,outfile=None,divide_by_height=False):
    labels = ['ML'+L+'_' for L in looks]
    print(labels)
    colordict={'GC1':'#326fd1',
               'GC2':'#32d15f',
               'GC3':'#d1c132',
               'GC4': '#d14732' 
              } # This is kind of random
    fig,axs=plt.subplots(1,2,figsize=(10,5))
    for i in range(len(looks)):
        label = labels[i]
        color=colordict.get('GC'+str(i+1))
        data = [f for f in tdx_data if label in f][0]
        x, y = plot_df['ChaveAGB_Change19-20'], get_phase_height_change(data,offset=offset)
        if divide_by_height == True:
            y = y / plot_df['MCH'] # scale according to height?
        slope, intercept, r, p, se = stats.linregress(x,y)
        label = label+'   r = '+'{:.2f}'.format(r) +'   p = '+'{:.2f}'.format(p)
        xx = np.linspace(x.min(),0,10)
        axs[0].plot(xx,slope*xx+intercept,linestyle='--',color=color)
        axs[0].plot(x,y,'o',label=label,color=color)
        #
        x = plot_df['ChaveAGB_pctChange19-20']
        xx = np.linspace(x.min(),0,10)
        slope, intercept, r, p, se = stats.linregress(x,y)
        axs[1].plot(xx,slope*xx+intercept,linestyle='--',color=color)
        axs[1].plot(x,y,'o',label=label,color=color)
    #plt.xlim(right=5)
    for ax in axs:
        ax.legend()
        ax.axvline(x=0,color='black')
        ax.axhline(y=0,color='black')
        ax.set_ylabel('$\Delta h_{\phi}$ (m)')
    axs[0].set_xlabel('$\Delta$ AGB (tonnes)')
    axs[1].set_xlabel('$\Delta$ AGB (%)')
    axs[0].set_xlim(right=5)
    axs[1].set_xlim(right=2)
    plt.legend()
    plt.tight_layout()
    if outfile != None:
        plt.savefig(outfile,dpi=400)
    plt.show()
```

```python
def height_change_map(f,offset=True):
    ds = open_ds(f)
    looks = int( f.split('_ML')[1].split('_')[0] )
    print('Looks: '+str(looks))
    if 'GAB' in f:
        country = 'G'
        df = g_df
    else:
        country = 'P'
        df = p_df

    if offset==True:
        offset = ds.mean(dim=['lat','lon'])
        print('Offset: ')
        print(offset.height.values)
        ds = ds - offset
    elif offset=='coherent':
        offset = ds.height.where( ds.coh > ds.coh.quantile(0.9,dim=['lat','lon']) ).median(dim=['lat','lon'])
        print('Coherent offset: ')
        print(offset.values)
#     with open_ds(get_reference(f,country)) as ref:
#         ds = ds.interp_like(ref,method='nearest')
#     results = {}
    timeseries = ds.height
    logdate = logging.get(country+'C1')[0]
    before = timeseries.sel(t=slice(None,logdate))
    after = timeseries.sel(t=slice(logdate,None))
    difference = after.mean(dim='t') - before.mean(dim='t')
    difference.plot(vmin=-20,vmax=20)
    plt.title('phase height change')
    plt.show()
    return difference
```

```python
def combined_asc_des():
    ref = open_ds('/disk/scratch/local.4/harry/DEM/NC/merged/subset/TDX_DEM_GAB_SM_A_ML4_merged_subset_negatives_fixed_drop6Nov.nc')
    asc = open_ds(f_gabA[0]).interp_like(ref).height
    des = open_ds(f_gabD[0]).interp_like(ref).height
    country = 'G'
    plotnames = [country+p for p in ['C1','C2','C3','C4']]

    asc = asc - asc.median(dim=['lat','lon'])
    des = des - des.median(dim=['lat','lon']) 
    results = {}
    for P in plotnames:
        A = crop(asc,P).mean(dim=['lat','lon'])
        D = crop(des,P).mean(dim=['lat','lon'])
        A = A - A.isel(t=0) # Set relative to first in A timeseries...
        D = D - D.isel(t=0) # Set relative to first in D timeseries
        timeseries = xr.concat([A,D],dim='t').sortby('t')
        logdate = logging.get(P)[0]
        before = timeseries.sel(t=slice(None,logdate))
        after = timeseries.sel(t=slice(logdate,None))
        difference = after.mean() - before.mean()
        results[P] = float(difference.values)
    return pd.Series(data=results)
```
