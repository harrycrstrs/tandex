```python
# Convenience routines ----------

def open_ds(dataset,looks):
    # Open net_cdf file ready to be cropped using rio
    filename = [x for x in dataset if '_ML'+str(looks)+'_' in x][0]
    try:
        ds = xr.open_dataset(filename).drop(['metadata','crs']).rio.write_crs('EPSG:4326')
    except:
        ds = xr.open_dataset(filename).rio.write_crs('EPSG:4326')
    return ds

def crop(ds,plot,all_touched=False):
    # Crop an xarray object 'ds' to plot
    # plot is a string e.g. 'PC1' or 'GC4'
    # pass all_touched = True to include pixels that partially intersect the plot
    ds.rio.set_spatial_dims(x_dim='lon', y_dim='lat', inplace=True)
    if 'B' in plot:
        shape = shps_bg[shps_bg.layer==plot].geometry
    else:
        shape = shps[shps.layer==plot].geometry
    return ds.rio.clip(shape,all_touched=all_touched)
```

```python
def get_coherent_points(ds, C=0.88):
    c = ds.coh.isel(t=0) > C
    for t in range(1,ds.t.size):
        c_t = ds.coh.isel(t=t) > C
        c = c & c_t
    print('Number of coherent points: ',float(c.sum()))
    return c

def do_coherent_offset(ds,C=0.88):
    coherent_points = get_coherent_points(ds,C=C)
    offset = ds.height.where( coherent_points ).mean(dim=['lat','lon'])
    print(offset.values)
    ds['height'] = ds.height - offset
    return ds
```

```python
def get_coh_changeMap(asc_offset,des_offset,show=False):
    cohA = asc_offset.coh.median(dim='t')
    cohD = des_offset.coh.median(dim='t')
    whereA = cohA > cohD
    if show==True:
        #whereA.rio.to_raster('/home/s1332488/chapter1/coherenceChoice.tif',dtype=np.int8)
        plt.figure(figsize=(12,12))
        plt.imshow(whereA)
        plt.savefig('/home/s1332488/chapter1/coherenceChoice.png',dpi=300)
        plt.show()
    neither = (cohA < 0.5) & (cohD < 0.5)
    #neither.plot()
    #plt.show()
    # hard code for now the before / after images
    changeA = asc_offset.height.isel(t=[2,3]).mean(dim='t') - asc_offset.height.isel(t=[0,1]).mean(dim='t')
    changeD = des_offset.height.isel(t=[2,3]).mean(dim='t') - des_offset.height.isel(t=[0,1]).mean(dim='t')
    change = changeA.where(whereA,other=changeD)
    return change.where(~neither)

#changeMap = get_coh_changeMap(do_coherent_offset(ascending,C=0.85),
#                              do_coherent_offset(descending,C=0.85))
#changeMap.rio.to_raster('/home/s1332488/chapter1/doublePassChangeMap_10looks.tif')
```
