```python
import xarray as xr
import os 
import glob
from matplotlib import pyplot as plt
import numpy as np
import rioxarray as rx
import geopandas as gpd
from datetime import datetime as dt
import pandas as pd
```

```python
def peru(list_of_strings):
    return [i for i in list_of_strings if 'PER' in i]

def gabon(list_of_strings):
    return [i for i in list_of_strings if 'GAB' in i]
```

```python
def get_orbit_dir(t):
    """
    t - value along time dimension of dataset
    returns 'A' or 'D' for ascending/descending orbit
    """
    md = '/home/s1332488/TDX/metadata.csv'
    date = np.datetime_as_string(t.values,unit='D')
    df = pd.read_csv(md)
    direction = df.query("date==@date")['orbit_direction'].values[0]
    return direction
```

```python
data_dir = '/disk/scratch/local/harry/temp/DEMs_merged_clean/'
os.chdir(data_dir)
files = glob.glob('*.nc')
f_per = peru(files)
f_gab = gabon(files)
```

```python
shp_dir='/exports/csce/datastore/geos/groups/MitchardGroupData/FODEX/core_plot_shapefiles/'
shps = glob.glob(shp_dir+'*.shp')
shp_per = [gpd.read_file(s) for s in peru(shps)]
shp_gab = [gpd.read_file(s) for s in gabon(shps)]
runway = gpd.read_file([s for s in shps if 'runway' in s][0])
```

```python
def plot_changes_gab(f):
    outname = f.split('DEM_')[1].split('_unwS')[0]+'_change.png'
    ds = xr.open_dataset(f).drop(['metadata','crs']).rio.write_crs('EPSG:4326').isel(t=slice(1,None))
    plots = [ds.rio.clip(s.geometry) for s in shp_gab]
    rwy = ds.rio.clip(runway.geometry)
    offset = rwy.height_corrected.mean(dim=['lat','lon'])
    #offset=ds.height_corrected.median(dim=['lat','lon'])
    fig,axs = plt.subplots(2,2,figsize=(10,10),sharex=True,sharey=True)
    for i in range(4):
        ax = axs.reshape(-1)[i]
        title = gabon(shps)[i].split('/')[-1].split('.shp')[0]
        if 'C1' in title:
            logging = ['2020-01-24','2020-01-27']
        elif 'C2' in title:
            logging = ['2020-01-27']
        elif 'C3' in title:
            logging = ['2020-01-24','2020-02-04']
        elif 'C4' in title:
            logging = ['2020-01-28']

        orbits = [get_orbit_dir(t) for t in ds.t]
        timeseries = (plots[i].height_corrected - offset).mean(dim=['lat','lon'])
        ascending = timeseries.isel(t= [o =='A' for o in orbits])
        descending = timeseries.isel(t= [o=='D' for o in orbits])
        ascending.plot(marker='x',ax=ax)
        descending.plot(marker='o',ax=ax)

        for date in logging:
            ax.axvline(dt.strptime(date,'%Y-%m-%d'),color='black',linestyle='--')
        ax.set_title(title)
    plt.savefig('/home/s1332488/TDX/results/'+outname,dpi=300)
    plt.show()
```

```python
for f in f_gab:
    plot_changes_gab(f)
```

```python
def map_changes_gab(f):
    outname = f.split('DEM_')[1].split('_unwS')[0]+'_changemap.png'
    ds = xr.open_dataset(f).drop(['metadata','crs']).rio.write_crs('EPSG:4326').isel(t=slice(1,None))
    #plots = [ds.rio.clip(s.geometry) for s in shp_gab]
    rwy = ds.rio.clip(runway.geometry)
    offset = rwy.height_corrected.mean(dim=['lat','lon'])
    #offset=ds.height_corrected.median(dim=['lat','lon'])
    timeseries = (ds.height_corrected - offset)
    descending = timeseries.isel(t= [o=='D' for o in [get_orbit_dir(t) for t in ds.t]])
    fig,axs = plt.subplots(2,2,figsize=(10,10),sharex=True,sharey=True)
    for i in range(3):
        ax = axs.reshape(-1)[i]
        for p in shp_gab:
            p.boundary.plot(ax=ax,color='red')
        j = i +1
        change = descending.isel(t=j) - descending.isel(t=i)
        change = change.coarsen(lat=10,boundary='trim').mean().coarsen(lon=10,boundary='trim').mean()
        change.plot(ax=ax,vmin=-5,vmax=5)
        ax.set_title(str(j)+' minus '+str(i))
    plt.savefig('/home/s1332488/TDX/results/'+outname,dpi=300)
    plt.show()
```
